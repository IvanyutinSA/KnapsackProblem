## Dimension: 10

### Random Multiple Attempts

```python
from tests.randomtests import Tests
from visual.auxiliary import display_pairs

tests = Tests(alg=Genetic(maximum_iterations=200))

items = 10
attempts = 30
y_hat, y, gen_times, lib_times = tests.RandomMultipleAttempts(items=items, attempts=attempts)

print("our\tlib")
display_pairs(y_hat, y)
```

```python
vis.ResultPlot(y_hat, y)
vis.MatchPlot(y_hat, y,)
```

```python
display_pairs(gen_times, lib_times)
vis.Time_Grafik(gen_times, lib_times)
```
